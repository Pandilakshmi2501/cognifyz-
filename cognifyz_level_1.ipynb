```python
import pandas as pd
df=pd.read_csv("Dataset .csv")
df.head()
```

Task - 1
Top 3 Cuisines

```python
top3 = df['Cuisines'].value_counts().head(3)
print(top3)
print("\nPercentage:\n", (top3/len(df))*100)
```

Task-2  City Analysis

```python
print("City with most restaurants:", df['City'].value_counts().idxmax())

city_ratings = df.groupby('City')['Aggregate rating'].mean()
print("\nAverage rating per city:\n", city_ratings)

print("\nCity with highest average rating:", city_ratings.idxmax())
```

Task-3 Price Range Distribution

```python
# Bar chart
df['Price range'].value_counts().plot(kind='bar',color='blue')
```

```python
print("\nPrice Range Percentage:\n", (df['Price range'].value_counts()/len(df))*100)
```

Task-4 Online Delivery

```python
delivery_percent = (df['Has Online delivery'].value_counts(normalize=True)*100)
print("\nOnline delivery %:\n", delivery_percent)

print("\nAvg rating with/without delivery:\n", df.groupby('Has Online delivery')['Aggregate rating'].mean())
```
